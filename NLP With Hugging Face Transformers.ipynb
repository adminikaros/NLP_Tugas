```python
# Import the necessary libraries
from transformers import pipeline
```

```python
# Zero-shot classification
classifier = pipeline("zero-shot-classification")
result = classifier(
    "This workshop focuses on deep learning and neural networks.",
    candidate_labels=["education", "technology", "healthcare"],
)
print(result)
```

```python
# Text generation with default model
generator = pipeline("text-generation")
gen_result = generator("In this workshop, participants will learn how to")
print(gen_result)
```

```python
# Text generation with a specific model (distilgpt2)
generator = pipeline("text-generation", model="distilgpt2")
gen_result = generator(
    "In this workshop, participants will learn how to",
    max_length=30,
    num_return_sequences=2,
)
print(gen_result)
```

```python
# Fill-mask task
unmasker = pipeline("fill-mask")
mask_result = unmasker("This workshop will teach you all about <mask> networks.", top_k=2)
print(mask_result)
```

```python
# Named entity recognition
ner = pipeline("ner", grouped_entities=True)
ner_result = ner("My name is John and I work at a tech startup in San Francisco.")
print(ner_result)
```

```python
# Question answering
qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
context = "Kita dapat menggunakan jaringan saraf untuk berbagai aplikasi, termasuk pengenalan suara dan visi komputer."
question = "Apa yang bisa digunakan jaringan saraf untuk?"
result = qa_pipeline(question=question, context=context)
print(f"Jawaban: {result['answer']}")
```

```python
# Sentiment analysis
classifier = pipeline("sentiment-analysis")
sentiment_result = classifier("I am excited to learn more about machine learning!")
print(sentiment_result)
```

```python
# Summarization
summarizer = pipeline("summarization")
summary_result = summarizer(
    """
    Over the years, the field of artificial intelligence has seen remarkable advancements.
    Technologies such as machine learning and natural language processing have revolutionized
    industries ranging from healthcare to finance. As these technologies continue to evolve,
    they enable more sophisticated applications, enhancing productivity and driving innovation.
    However, ethical considerations and the impact of AI on society must also be addressed
    to ensure responsible development and deployment of these technologies.
    """
)
print(summary_result)
```

```python
# Translation from Indonesian to English
translator = pipeline("translation_id_to_en", model="Helsinki-NLP/opus-mt-id-en")
text_to_translate = "Hari ini saya belajar tentang kecerdasan buatan."
translation_result = translator(text_to_translate)
print(translation_result[0]['translation_text'])
```

```python
# Question answering again
qa_pipeline = pipeline("question-answering")
context = "Kita dapat menggunakan jaringan saraf untuk berbagai aplikasi, termasuk pengenalan suara dan visi komputer."
question = "Kita dapat menggunakan jaringan saraf untuk berbagai aplikasi termasuk?"
result = qa_pipeline(question=question, context=context)
print(f"Jawaban: {result['answer']}")
```
